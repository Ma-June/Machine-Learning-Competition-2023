```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(3,3))
plt.hist(df.GR.values)
plt.xlabel('GR')
plt.ylabel('Frequency')
plt.title('Gamma-ray record in Train #1')
plt.grid('on')

plt.savefig("figure_1.png")
```

```python
plt.figure(figsize=(3,10))
plt.plot(df.GR.values, df.DEPT.values)
plt.gca().invert_yaxis()
plt.title('GR')
plt.ylabel('DEPT')
plt.xlabel('Unit: API')
plt.grid('on')
```

```python
plt.figure(figsize=(12,10))
plt.suptitle('Training data #1')

plt.subplot(1,4,1)
plt.plot(df.GR.values, df.DEPT.values)
plt.gca().invert_yaxis()
plt.title('Gamma-ray')
plt.ylabel('DEPT')
plt.xlabel('Unit: API')
plt.grid('on')

plt.subplot(1,4,2)
plt.plot(df.RHOB.values, df.DEPT.values)
plt.gca().invert_yaxis()
plt.title('density')
plt.ylabel('DEPT')
plt.xlabel('Unit: g/cc')
plt.xlim([2,2.75])
plt.grid('on')

plt.subplot(1,4,3)
plt.plot(df.NPHI.values, df.DEPT.values)
plt.gca().invert_yaxis()
plt.title('neutron')
plt.ylabel('DEPT')
plt.xlabel('Unit: %')
plt.grid('on')

plt.subplot(1,4,4)
plt.plot(df.RD.values, df.DEPT.values)
plt.gca().invert_yaxis()
plt.title('resistivity')
plt.ylabel('DEPT')
plt.xlabel('Unit: o')
plt.xlim([0,25])
plt.grid('on')

plt.tight_layout()
```

```python
for i in range(1,10):
    plt.figure(figsize=(12,10))
    plt.suptitle(f'Training data # {i}')
    df = pd.read_csv(f"data/train/aligned_well_0{i}.csv")

    plt.subplot(1,4,1)
    plt.plot(df.GR.values, df.DEPT.values)
    plt.gca().invert_yaxis()
    plt.title('Gamma-ray')
    plt.ylabel('DEPT')
    plt.xlabel('Unit: API')
    plt.grid('on')

    plt.subplot(1,4,2)
    plt.plot(df.RHOB.values, df.DEPT.values)
    plt.gca().invert_yaxis()
    plt.title('density')
    plt.ylabel('DEPT')
    plt.xlabel('Unit: g/cc')
    plt.xlim([2,2.75])
    plt.grid('on')

    plt.subplot(1,4,3)
    plt.plot(df.NPHI.values, df.DEPT.values)
    plt.gca().invert_yaxis()
    plt.title('neutron')
    plt.ylabel('DEPT')
    plt.xlabel('Unit: %')
    plt.grid('on')

    plt.subplot(1,4,4)
    plt.plot(df.RD.values, df.DEPT.values)
    plt.gca().invert_yaxis()
    plt.title('resistivity')
    plt.ylabel('DEPT')
    plt.xlabel('Unit: o')
    plt.xlim([0,25])
    plt.grid('on')

    plt.tight_layout()

plt.savefig("figure_2.png")
```
